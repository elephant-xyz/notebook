```python
# @title Step 2: Prepare, Transform, Validate, Hash & Upload
# @title Step 0: Get Next Uncompleted Parcel
# ===========================
# Imports & Setup
# ===========================
import os
import re
import json
import csv
import time
import shutil
import zipfile
import tempfile
import subprocess
from pathlib import Path
from dataclasses import dataclass
from typing import Any, NotRequired, TypedDict, cast

from google.oauth2.service_account import Credentials
import gspread
import requests

from requests import Session
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry
from urllib.parse import urlencode, urlparse, urlunparse, parse_qsl, quote

from dotenv import load_dotenv
from IPython.display import HTML, display

os.environ["PINATA_JWT"] = "eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJ1c2VySW5mb3JtYXRpb24iOnsiaWQiOiIyOGIxOWFiMy1hZGFiLTRmOWEtOTUzNi1mYmRjN2YyZjU3ZGUiLCJlbWFpbCI6Im15a3l0YS5vdnNpaWVua28rYnJlYXZhcmRAc3RhaXJjYXNlLmNvIiwiZW1haWxfdmVyaWZpZWQiOnRydWUsInBpbl9wb2xpY3kiOnsicmVnaW9ucyI6W3siZGVzaXJlZFJlcGxpY2F0aW9uQ291bnQiOjEsImlkIjoiRlJBMSJ9LHsiZGVzaXJlZFJlcGxpY2F0aW9uQ291bnQiOjEsImlkIjoiTllDMSJ9XSwidmVyc2lvbiI6MX0sIm1mYV9lbmFibGVkIjpmYWxzZSwic3RhdHVzIjoiQUNUSVZFIn0sImF1dGhlbnRpY2F0aW9uVHlwZSI6InNjb3BlZEtleSIsInNjb3BlZEtleUtleSI6IjI2YmIwZmNiNDY0N2E2NjE3NWY0Iiwic2NvcGVkS2V5U2VjcmV0IjoiMWNiZjEzMWNlNjcwZTk1ZTJmNjkyYjk2NDhhN2RhNWE1Y2FiODgyMjIxMWRkNjk5MWVkNjYzMzM1ZDc1Zjg0NyIsImV4cCI6MTc4OTIxNzcwMX0.__W89NY9Eic3NMNQWpirhSI7DyT_etgQjsOqdLaJS1I"


# Install Playwright + deps for fallback browser flow
!pip install playwright==1.46.0 python-dotenv -q > /dev/null
!playwright install-deps > /dev/null 2>&1
!python -m playwright install chromium > /dev/null 2>&1

# ===========================
# Constants
# ===========================
SERVICE_ACCOUNT_FILE = "service-account.json"

# Sheet with parcels + propertyCid (seed hash)
SPREADSHEET_ID = "1o1JscCqPNsBItQzBkVnD-VWqlZwNl_Z92JfWg1POYSE"

# External scripts zip (always from S3)
LEON_SCRIPTS_URL = "https://duna-seeds.s3.us-east-1.amazonaws.com/leon.zip"
LEON_SCRIPTS_ZIP = "leon.zip"

# Ensure service account JSON is present (download from S3 if needed)
if not os.path.exists(SERVICE_ACCOUNT_FILE):
    resp = requests.get("https://duna-seeds.s3.us-east-1.amazonaws.com/service-account.json", timeout=60)
    resp.raise_for_status()
    with open(SERVICE_ACCOUNT_FILE, "wb") as f:
        f.write(resp.content)

# ===========================
# UI Helpers
# ===========================
def error_box(message: str):
    display(HTML(f"""
        <div style="
            background:#ffebee;
            color:#b71c1c;
            padding:12px;
            border-left:5px solid #c62828;
            margin:10px 0;
            border-radius:4px;
            font-size:15px;">
            ‚ùå <b>{message}</b>
        </div>
    """))

def success_box(message: str):
    display(HTML(f"""
        <div style="
            background:#e8f5e9;
            color:#1b5e20;
            padding:12px;
            border-left:5px solid #2e7d32;
            margin:10px 0;
            border-radius:4px;
            font-size:15px;">
            ‚úÖ <b>{message}</b>
        </div>
    """))

# ===========================
# Types for Prepare / Packager
# ===========================
class SourceHttpRequest(TypedDict, total=False):
    method: str
    url: str
    multiValueQueryString: dict[str, Any]
    headers: dict[str, str]
    body: Any
    json: Any

class PropertySeed(TypedDict, total=False):
    parcel_id: NotRequired[str]
    request_identifier: NotRequired[str]
    source_http_request: SourceHttpRequest

@dataclass(frozen=True)
class PackagerConfig:
    input_zip: Path
    output_zip: Path
    timeout_sec: float = 30.0
    retries: int = 3
    backoff_factor: float = 1.6
    status_forcelist: tuple[int, ...] = (429, 500, 502, 503, 504)
    enable_browser_fallback: bool = True

# ===========================
# PropertyZipPackager
# ===========================
class PropertyZipPackager:
    """
    Reads prepare-input.zip:
        prepare-input.zip
        ‚îî‚îÄ‚îÄ seed_output/
            ‚îú‚îÄ‚îÄ property_seed.json
            ‚îî‚îÄ‚îÄ unnormalized_address.json

    Uses property_seed.json.source_http_request to fetch the
    parcel HTML/JSON and writes it back into seed_output/,
    then zips that as output.zip (which we then re-pack to prepared-county.zip).
    """
    def __init__(self, config: PackagerConfig, session: Session | None = None) -> None:
        self.cfg = config
        self.session = session or self._build_session()

    def run(self) -> Path:
        self._validate_paths()
        with tempfile.TemporaryDirectory(prefix="prop_zip_packager_") as tmpdir:
            workdir = Path(tmpdir)
            self._extract_input_zip(workdir)

            seed_dir = workdir / "seed_output"
            seed = self._load_property_seed(seed_dir / "property_seed.json")
            body_text = self._fetch_html_from_seed(seed)
            file_stem = self._resolve_identifier(seed)

            # Decide whether it's JSON or HTML
            try:
                parsed = json.loads(body_text)
                out_path = seed_dir / f"{file_stem}.json"
                out_path.write_text(json.dumps(parsed, ensure_ascii=False, indent=2), encoding="utf-8")
            except json.JSONDecodeError:
                out_path = seed_dir / f"{file_stem}.html"
                out_path.write_text(body_text, encoding="utf-8")

            self._write_output_zip(workdir)

        return self.cfg.output_zip

    def _build_session(self) -> Session:
        retry = Retry(
            total=self.cfg.retries,
            connect=self.cfg.retries,
            read=self.cfg.retries,
            status=self.cfg.retries,
            backoff_factor=self.cfg.backoff_factor,
            status_forcelist=self.cfg.status_forcelist,
            allowed_methods={"GET", "POST"},
            raise_on_status=False,
            respect_retry_after_header=True,
        )
        adapter = HTTPAdapter(max_retries=retry)
        s = requests.Session()
        s.headers.update({
            "User-Agent": (
                "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                "AppleWebKit/537.36 Chrome/124.0 Safari/537.36"
            )
        })
        s.mount("http://", adapter)
        s.mount("https://", adapter)
        return s

    def _validate_paths(self) -> None:
        if not self.cfg.input_zip.exists():
            raise FileNotFoundError(f"Input ZIP not found: {self.cfg.input_zip}")
        self.cfg.output_zip.parent.mkdir(parents=True, exist_ok=True)

    def _extract_input_zip(self, workdir: Path) -> None:
        with zipfile.ZipFile(self.cfg.input_zip, "r") as zf:
            zf.extractall(workdir)
        seed_dir = workdir / "seed_output"
        if not seed_dir.is_dir():
            raise ValueError("Input ZIP must contain a 'seed_output/' directory.")
        if not (seed_dir / "property_seed.json").is_file():
            raise ValueError("Missing 'seed_output/property_seed.json' in input ZIP.")

    def _load_property_seed(self, path: Path) -> PropertySeed:
        try:
            data = json.loads(path.read_text(encoding="utf-8"))
        except json.JSONDecodeError as e:
            raise ValueError(f"Invalid JSON in {path}: {e}") from e
        if not isinstance(data, dict) or "source_http_request" not in data:
            raise ValueError("property_seed.json must include 'source_http_request'")
        return cast(PropertySeed, data)

    def _resolve_identifier(self, seed: PropertySeed) -> str:
        pid = (seed.get("parcel_id") or "").strip()
        rid = (seed.get("request_identifier") or "").strip()
        ident = pid or rid
        if not ident:
            raise ValueError("property_seed.json must include 'parcel_id' or 'request_identifier'")
        return ident

    def _normalize_query_params(self, mvqs: dict[str, Any] | None) -> dict[str, Any]:
        if not mvqs:
            return {}
        params: dict[str, Any] = {}
        for k, v in mvqs.items():
            params[k] = v[0] if isinstance(v, list) and v else v
        return params

    def _build_url(self, base: str, params: dict[str, Any]) -> str:
        if not params:
            return base
        parsed = urlparse(base)
        existing = dict(parse_qsl(parsed.query, keep_blank_values=True))
        existing.update({k: "" if v is None else str(v) for k, v in params.items()})
        new_qs = urlencode(existing, doseq=True, quote_via=quote)
        return urlunparse(parsed._replace(query=new_qs))

    def _fetch_html_from_seed(self, seed: PropertySeed) -> str:
        src = seed.get("source_http_request", {})
        method = (src.get("method") or "GET").upper()
        url = src.get("url")
        if not url:
            raise ValueError("source_http_request.url is required")

        params = self._normalize_query_params(
            cast(dict[str, Any] | None, src.get("multiValueQueryString"))
        )
        headers = cast(dict[str, str] | None, src.get("headers")) or {}
        body = src.get("body")
        json_body = src.get("json")

        headers.setdefault("Accept", "application/json, text/html;q=0.9")
        kwargs: dict[str, Any] = {"timeout": self.cfg.timeout_sec, "headers": headers}

        if method == "GET":
            kwargs["params"] = params
        elif method == "POST":
            kwargs["params"] = params
            if json_body is not None:
                kwargs["json"] = json_body
            elif body is not None:
                kwargs["data"] = body
        else:
            raise ValueError(f"Unsupported HTTP method: {method}")

        resp = self.session.request(method, url, **kwargs)
        resp.raise_for_status()
        encoding = resp.encoding or resp.apparent_encoding
        html_or_json = resp.content.decode(encoding or "utf-8", errors="replace")

        # If JSON, just return as-is
        if "application/json" in (resp.headers.get("Content-Type") or ""):
            return html_or_json

        # Fallback to browser (Playwright) if there's a "Continue" modal
        if self.cfg.enable_browser_fallback and self._looks_like_continue_modal(html_or_json):
            final_html = self._fetch_via_browser_click_continue(
                full_url=self._build_url(url, params),
                timeout=self.cfg.timeout_sec
            )
            if final_html:
                return final_html

        return html_or_json

    def _looks_like_continue_modal(self, html: str) -> bool:
        h = html.lower()
        return (
            ("id=\"pnlissues\"" in h) or ("id='pnlissues'" in h)
        ) and (
            "btncontinue" in h or "value=\"continue\"" in h or ">continue<" in h
        )

    def _fetch_via_browser_click_continue(self, full_url: str, timeout: float) -> str | None:
        try:
            import asyncio, threading, queue
            from playwright.async_api import async_playwright, TimeoutError as PWTimeout
        except Exception:
            return None

        t_ms = int(max(2000, timeout * 1000))
        issue_panel_sel = "#pnlIssues"
        continue_selectors = [
            '#pnlIssues input[name="btnContinue"]',
            '#pnlIssues input[value="Continue"]',
            '#pnlIssues button[value="Continue"]',
            '#btnContinue',
        ]
        property_markers = [
            "#parcelLabel",
            ".sectionTitle",
            "table.detailsTable",
            ".textPanel",
            "[id*='Property']",
        ]

        async def is_visible(el) -> bool:
            try:
                box = await el.bounding_box()
                if not box:
                    return False
                disp = await el.evaluate("el => getComputedStyle(el).display")
                vis = await el.evaluate("el => getComputedStyle(el).visibility")
                return disp != "none" and vis != "hidden"
            except Exception:
                return False

        async def do_flow() -> str | None:
            async with async_playwright() as pw:
                browser = await pw.chromium.launch(
                    headless=True,
                    args=["--no-sandbox", "--disable-dev-shm-usage"]
                )
                context = await browser.new_context(
                    user_agent=(
                        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                        "AppleWebKit/537.36 Chrome/124.0 Safari/537.36"
                    ),
                    viewport={"width": 1280, "height": 2000},
                    locale="en-US",
                )
                page = await context.new_page()
                await page.goto(full_url, wait_until="networkidle", timeout=t_ms)

                modal_present = False
                try:
                    await page.wait_for_selector(issue_panel_sel, state="visible", timeout=8000)
                    modal_present = True
                except PWTimeout:
                    pass

                async def find_in_any_frame(selectors):
                    for fr in page.context.pages[0].frames:
                        for sel in selectors:
                            try:
                                el = await fr.query_selector(sel)
                                if el and await is_visible(el):
                                    return fr, el, sel
                            except Exception:
                                pass
                    return None, None, None

                clicked = False
                fr, btn, _ = await find_in_any_frame(continue_selectors)
                if btn:
                    try:
                        await btn.click(timeout=min(5000, t_ms))
                        clicked = True
                    except Exception:
                        try:
                            await fr.evaluate("(el)=>el.click()", btn)
                            clicked = True
                        except Exception:
                            pass

                if not clicked:
                    try:
                        has_postback = await page.evaluate(
                            "() => typeof window.__doPostBack === 'function'"
                        )
                        if has_postback:
                            await page.evaluate("__doPostBack('btnContinue','')")
                            clicked = True
                    except Exception:
                        pass

                if clicked or modal_present:
                    try:
                        await page.wait_for_selector(
                            issue_panel_sel,
                            state="hidden",
                            timeout=min(15000, t_ms)
                        )
                    except PWTimeout:
                        for sel in property_markers:
                            try:
                                await page.wait_for_selector(sel, state="visible", timeout=4000)
                                break
                            except PWTimeout:
                                continue

                html = await page.content()
                await context.close()
                await browser.close()
                return html

        try:
            loop = asyncio.get_event_loop()
            if loop.is_running():
                q = queue.Queue()

                def runner():
                    new_loop = asyncio.new_event_loop()
                    asyncio.set_event_loop(new_loop)
                    try:
                        q.put(new_loop.run_until_complete(do_flow()))
                    except Exception:
                        q.put(None)
                    finally:
                        new_loop.close()

                t = threading.Thread(target=runner, daemon=True)
                t.start()
                t.join()
                return q.get()
            else:
                return loop.run_until_complete(do_flow())
        except RuntimeError:
            return asyncio.run(do_flow())
        except Exception:
            return None

    def _write_output_zip(self, workdir: Path) -> None:
        src_dir = Path(workdir) / "seed_output"
        tmp_path = self.cfg.output_zip.with_suffix(self.cfg.output_zip.suffix + ".tmp")
        with zipfile.ZipFile(tmp_path, "w", compression=zipfile.ZIP_DEFLATED) as zf:
            for p in sorted(src_dir.rglob("*")):
                if p.is_file():
                    zf.write(p, p.relative_to(workdir).as_posix())
        shutil.move(tmp_path, self.cfg.output_zip)

# ===========================
# Helpers: CSV / CLI
# ===========================
def has_submit_errors(path="submit_errors.csv") -> bool:
    try:
        with open(path, newline='', encoding='utf-8') as f:
            return next(csv.DictReader(f), None) is not None
    except FileNotFoundError:
        return False

def get_upload_results(path="county-results.csv") -> dict[str, Any]:
    with open(path, newline='', encoding='utf-8') as f:
        row = next(csv.DictReader(f), None)
        if row is None:
            raise ValueError("county-results.csv is empty")
        return row

def ensure_scripts_zip(local_path: str = LEON_SCRIPTS_ZIP) -> str:
    if os.path.exists(local_path):
        return local_path
    print(f"Downloading scripts from {LEON_SCRIPTS_URL} ...")
    resp = requests.get(LEON_SCRIPTS_URL, timeout=120)
    resp.raise_for_status()
    with open(local_path, "wb") as f:
        f.write(resp.content)
    print("‚úî Scripts zip downloaded")
    return local_path

def find_zip_file() -> str | None:
    """Find the most recent downloaded ZIP file."""
    import glob
    exclude = ["prepare-input.zip", "prepared-county.zip",
               "transformed-county.zip", "hashed-data.zip"]
    zip_files = [f for f in glob.glob("*.zip") if f not in exclude]
    if not zip_files:
        return None
    zip_files.sort(key=os.path.getmtime, reverse=True)
    return zip_files[0]

# ===========================
# NEW: Build prepare-input.zip from *new* parcel zip
# ===========================
def build_prepare_input_from_parcel_zip(parcel_zip_path: str):
    """
    Given the downloaded parcel zip in the *new* format, build prepare-input.zip
    in the format expected by PropertyZipPackager:

        prepare-input.zip
        ‚îî‚îÄ‚îÄ seed_output/
            ‚îú‚îÄ‚îÄ property_seed.json
            ‚îî‚îÄ‚îÄ unnormalized_address.json (if present)
    """
    temp_dir = "./temp_parcel_unzip"
    shutil.rmtree(temp_dir, ignore_errors=True)
    os.makedirs(temp_dir, exist_ok=True)

    # 1) Extract the new-format parcel zip
    with zipfile.ZipFile(parcel_zip_path, "r") as z:
        z.extractall(temp_dir)

    def find_in_temp(filename: str):
        for root, _, files in os.walk(temp_dir):
            if filename in files:
                return os.path.join(root, filename)
        return None

    # 2) Locate property_seed.json (required)
    prop_seed_path = find_in_temp("property_seed.json")
    if not prop_seed_path:
        shutil.rmtree(temp_dir, ignore_errors=True)
        raise FileNotFoundError("property_seed.json not found in parcel zip")

    # 3) Locate unnormalized_address.json (optional)
    unnorm_path = find_in_temp("unnormalized_address.json")

    # 4) Build a seed_output/ folder in temp and copy JSONs into it
    seed_output_dir = os.path.join(temp_dir, "seed_output")
    os.makedirs(seed_output_dir, exist_ok=True)

    shutil.copy(prop_seed_path, os.path.join(seed_output_dir, "property_seed.json"))
    if unnorm_path:
        shutil.copy(unnorm_path, os.path.join(seed_output_dir, "unnormalized_address.json"))

    # 5) Create prepare-input.zip with seed_output/ preserved
    if os.path.exists("prepare-input.zip"):
        os.remove("prepare-input.zip")

    with zipfile.ZipFile("prepare-input.zip", "w", zipfile.ZIP_DEFLATED) as zip_out:
        for root, _, files in os.walk(seed_output_dir):
            for f in files:
                full_path = os.path.join(root, f)
                # arcname should start with seed_output/...
                arcname = os.path.relpath(full_path, temp_dir)
                zip_out.write(full_path, arcname)

    shutil.rmtree(temp_dir, ignore_errors=True)
    print(f"‚úî Built prepare-input.zip")

# ===========================
# Step 0: Get Next Parcel (from Sheet) + Download + build prepare-input.zip
# ===========================
current_parcel: dict[str, Any] | None = None

def get_next_parcel():
    """
    - Reads Google Sheet
    - Finds first row with:
        parcel_id present
        Transaction ID empty
    - Also reads propertyCid (or seed hash) from the same row
    - Downloads the parcel zip
    - Builds prepare-input.zip from property_seed.json + unnormalized_address.json
    """
    global current_parcel

    creds = Credentials.from_service_account_file(
        SERVICE_ACCOUNT_FILE,
        scopes=["https://www.googleapis.com/auth/spreadsheets"]
    )
    gc = gspread.authorize(creds)
    spreadsheet = gc.open_by_key(SPREADSHEET_ID)
    worksheet = spreadsheet.sheet1

    all_values = worksheet.get_all_values()
    if len(all_values) < 2:
        return None

    headers = all_values[0]

    # Required columns
    try:
        parcel_col = headers.index("parcel_id")
        transaction_col = headers.index("Transaction ID")
    except ValueError:
        raise ValueError("Sheet must have 'parcel_id' and 'Transaction ID' columns")

    # Optional but really required for hash: propertyCid / seed hash
    property_cid_col = None
    for candidate in ["propertyCid", "property_cid", "seed_hash", "seedHash", "seed_cid"]:
        if candidate in headers:
            property_cid_col = headers.index(candidate)
            break

    if property_cid_col is None:
        print("‚ö† No propertyCid / seed_hash column found. Hash step will fail without it.")

    # Find first unprocessed row
    for row_idx in range(1, len(all_values)):
        row = all_values[row_idx]
        parcel_id = row[parcel_col] if parcel_col < len(row) else ""
        transaction_id = row[transaction_col] if transaction_col < len(row) else ""
        property_cid = (
            row[property_cid_col].strip()
            if property_cid_col is not None and property_cid_col < len(row)
            else ""
        )

        if not parcel_id:
            continue
        if transaction_id:
            # already processed
            continue

        # Get hyperlink (download URL) from the parcel_id cell (if it's a HYPERLINK formula)
        cell_address = gspread.utils.rowcol_to_a1(row_idx + 1, parcel_col + 1)
        cell_data = worksheet.acell(cell_address, value_render_option='FORMULA').value

        hyperlink_match = re.search(r'=HYPERLINK\("([^"]+)"', str(cell_data))
        if hyperlink_match:
            download_url = hyperlink_match.group(1)
        else:
            download_url = parcel_id if parcel_id.startswith("http") else None

        if not download_url:
            continue

        try:
            # Download the parcel zip
            resp = requests.get(download_url, timeout=120)
            resp.raise_for_status()

            content_disp = resp.headers.get('Content-Disposition', '')
            if 'filename=' in content_disp:
                filename = re.findall('filename=(.+)', content_disp)[0].strip('"')
            else:
                filename = download_url.split('/')[-1] or f"{parcel_id}.zip"

            with open(filename, 'wb') as f:
                f.write(resp.content)

            if not filename.lower().endswith(".zip"):
                raise ValueError(f"Downloaded file is not a ZIP: {filename}")

            # Build prepare-input.zip from the new-format parcel ZIP
            build_prepare_input_from_parcel_zip(filename)

            current_parcel = {
                "row_number": row_idx + 1,
                "parcel_id": parcel_id,
                "download_url": download_url,
                "filename": filename,
                "property_cid": property_cid,
            }

            return current_parcel

        except Exception as e:
            error_box(f"Failed to download or process parcel at row {row_idx+1}: {e}")
            return None

    # No more rows
    return None

# Run step 0 immediately so user sees the chosen parcel
get_next_parcel()

# ===========================
# Pipeline Steps
# ===========================
def run_prepare() -> bool:
    """
    Uses PropertyZipPackager on prepare-input.zip, then creates prepared-county.zip
    that contains:
      - ALL files from the original parcel ZIP (flattened so there's no top-level
        'data/' folder inside the ZIP)
      - PLUS the HTML/JSON fetched via property_seed.json (also at the root)

    While doing this, we also derive SEED_DATA_GROUP_ID from a "special" JSON file
    in the original parcel zip (typically the bafk...json), and store it in:

        - global SEED_DATA_GROUP_ID (Python)
        - os.environ["SEED_DATA_GROUP_ID"] (for CLI / Node scripts)

    Final layout inside prepared-county.zip (example):

        address.json
        parcel.json
        property_seed.json
        unnormalized_address.json
        address_has_parcel.json
        <data-group-id>.json        (e.g. bafkreihozy...json)
        <parcel_id>.html or .json   (from the packager)
    """
    global current_parcel, SEED_DATA_GROUP_ID

    try:
        if not current_parcel or "filename" not in current_parcel:
            error_box("No current parcel/filename. Run get_next_parcel() first.")
            return False

        parcel_zip_path = current_parcel["filename"]
        if not os.path.exists(parcel_zip_path):
            error_box(f"Parcel ZIP not found on disk: {parcel_zip_path}")
            return False

        # 1) Run packager: prepare-input.zip -> output.zip with seed_output/*
        cfg = PackagerConfig(
            input_zip=Path("prepare-input.zip"),
            output_zip=Path("output.zip"),
            timeout_sec=30.0,
            enable_browser_fallback=True,
        )
        packager = PropertyZipPackager(cfg)
        packager.run()

        # Temporary working dirs
        temp_scraped = "temp_scraped"      # unpack of output.zip (seed_output/*)
        temp_orig = "temp_orig_parcel"     # unpack of original parcel ZIP
        temp_final = "temp_final_prepare"  # everything merged here before zipping

        shutil.rmtree(temp_scraped, ignore_errors=True)
        shutil.rmtree(temp_orig, ignore_errors=True)
        shutil.rmtree(temp_final, ignore_errors=True)

        os.makedirs(temp_scraped, exist_ok=True)
        os.makedirs(temp_orig, exist_ok=True)
        os.makedirs(temp_final, exist_ok=True)

        # 2) Unzip the packager output (output.zip -> temp_scraped/seed_output/*)
        with zipfile.ZipFile("output.zip", "r") as z:
            z.extractall(temp_scraped)

        seed_output_dir = os.path.join(temp_scraped, "seed_output")
        if not os.path.isdir(seed_output_dir):
            raise RuntimeError("Packager output missing seed_output/ directory")

        # 3) Unzip the original parcel ZIP (new format) to temp_orig
        with zipfile.ZipFile(parcel_zip_path, "r") as z:
            z.extractall(temp_orig)

        # 4) Copy EVERYTHING from the original parcel zip into temp_final
        #    - If there's a top-level 'data' folder, copy its contents to the root of temp_final
        #    - Otherwise, copy whatever is at the root of temp_orig
        #    Also: derive SEED_DATA_GROUP_ID from a special JSON filename.
        SEED_DATA_GROUP_ID = None  # reset each run

        data_dir = os.path.join(temp_orig, "data")
        # Decide which dir to inspect/copy from
        source_root = data_dir if os.path.isdir(data_dir) else temp_orig

        # Files we consider "standard", not the data-group file
        ignore_names = {
            "address.json",
            "parcel.json",
            "property_seed.json",
            "unnormalized_address.json",
            "address_has_parcel.json",
        }

        # First pass: figure out SEED_DATA_GROUP_ID from a JSON filename
        # Heuristic: pick the first non-standard JSON as the data group id
        for name in os.listdir(source_root):
            lower = name.lower()
            if lower.endswith(".json") and name not in ignore_names:
                SEED_DATA_GROUP_ID = os.path.splitext(name)[0]
                os.environ["SEED_DATA_GROUP_ID"] = SEED_DATA_GROUP_ID  # export for CLI if needed
                print(f"‚úî SEED_DATA_GROUP_ID set to: {SEED_DATA_GROUP_ID}")
                break

        # Second pass: actually copy everything into temp_final
        if os.path.isdir(data_dir):
            # Mimic: mv data/* .
            for name in os.listdir(data_dir):
                src = os.path.join(data_dir, name)
                dst = os.path.join(temp_final, name)
                if os.path.isdir(src):
                    shutil.copytree(src, dst, dirs_exist_ok=True)
                else:
                    shutil.copy2(src, dst)
        else:
            # No data/ folder ‚Äì copy top-level files from temp_orig
            for name in os.listdir(temp_orig):
                src = os.path.join(temp_orig, name)
                dst = os.path.join(temp_final, name)
                if os.path.isdir(src):
                    shutil.copytree(src, dst, dirs_exist_ok=True)
                else:
                    shutil.copy2(src, dst)

        # 5) Copy HTML/JSON from seed_output (packager output) into temp_final root
        for name in os.listdir(seed_output_dir):
            src = os.path.join(seed_output_dir, name)
            if os.path.isfile(src) and (
                name.lower().endswith(".html")
                or name.lower().endswith(".json")
            ):
                dst = os.path.join(temp_final, name)
                shutil.copy2(src, dst)

        # 6) Zip up EVERYTHING in temp_final as prepared-county.zip
        if os.path.exists("prepared-county.zip"):
            os.remove("prepared-county.zip")

        shutil.make_archive(
            "prepared-county",
            "zip",
            root_dir=temp_final,  # cd temp_final
            base_dir="."          # zip everything at root (no extra folder)
        )

        # 7) Clean up
        shutil.rmtree(temp_scraped, ignore_errors=True)
        shutil.rmtree(temp_orig, ignore_errors=True)
        shutil.rmtree(temp_final, ignore_errors=True)
        os.remove("output.zip")

        print("‚úî Prepare done (prepared-county.zip now includes ALL original files + scraped HTML/JSON at root)")
        return True

    except Exception as e:
        error_box(f"Prepare failed: {e}")
        return False




def run_transform() -> bool:
    try:
        scripts_zip = ensure_scripts_zip()
        subprocess.run(
            [
                "npx", "-y", "@elephant-xyz/cli@latest", "transform",
                "--input-zip", "prepared-county.zip",
                "--output-zip", "transformed-county.zip",
                "--scripts-zip", scripts_zip,
            ],
            stdout=subprocess.DEVNULL,
            stderr=subprocess.PIPE,
            check=True,
            text=True,
        )
        print("‚úî Transform done")
        return True
    except subprocess.CalledProcessError as e:
        error_box(f"Transform failed (exit code {e.returncode})")
        return False

def run_validate() -> bool:
    try:
        subprocess.run(
            ["npx", "-y", "@elephant-xyz/cli@latest", "validate", "transformed-county.zip"],
            stdout=subprocess.DEVNULL,
            stderr=subprocess.PIPE,
            check=True,
            text=True,
        )
        if has_submit_errors():
            error_box("Validate failed, please check submit_errors.csv for details")
            return False
        print("‚úî Validate done")
        return True
    except subprocess.CalledProcessError:
        error_box("Validate failed, please check submit_errors.csv for details")
        return False

def run_hash() -> bool:
    global current_parcel
    try:
        if not current_parcel or not current_parcel.get("property_cid"):
            error_box(
                "No property CID / seed hash found for current parcel.\n"
                "Please ensure the Google Sheet row has a 'propertyCid' (or 'seed_hash') value."
            )
            return False

        seed_group_cid = current_parcel["property_cid"]
        subprocess.run(
            [
                "npx", "-y", "@elephant-xyz/cli@latest", "hash",
                "transformed-county.zip",
                "--output-zip", "hashed-data.zip",
                "--output-csv", "county-results.csv",
                "--property-cid", seed_group_cid,
            ],
            stdout=subprocess.DEVNULL,
            stderr=subprocess.PIPE,
            check=True,
            text=True,
        )
        if has_submit_errors():
            error_box("Hash failed, please check submit_errors.csv for details")
            return False
        print("‚úî Hash done")
        return True
    except subprocess.CalledProcessError as e:
        error_box(f"Hash failed (exit code {e.returncode})")
        return False

def run_upload() -> bool:
    try:
        subprocess.run(
            ["npx", "-y", "@elephant-xyz/cli@1.28.4", "upload", "hashed-data.zip"],
            stdout=subprocess.DEVNULL,
            stderr=subprocess.PIPE,
            check=True,
            text=True,
        )
        upload_results = get_upload_results()
        data_cid, html_link = upload_results["dataCid"], upload_results["htmlLink"]
        print("‚úî Upload done")
        print("dataCid:", data_cid)
        print("htmlLink:", html_link)
        return True
    except subprocess.CalledProcessError as e:
        error_box(f"Upload failed (exit code {e.returncode})")
        return False

# ===========================
# Run All
# ===========================
def run_all():
    if not run_prepare():
        return
    if not run_transform():
        return
    if not run_validate():
        return
    if not run_hash():
        return
    if not run_upload():
        return
    success_box("Done, you can proceed to the next step")

# Actually run the pipeline
run_all()
```
